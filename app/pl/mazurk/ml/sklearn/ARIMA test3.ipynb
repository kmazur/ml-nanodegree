```python
from arima import predictARIMA
from datetime import datetime
from StockDataSource import YahooDataSource

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sns


%matplotlib inline

yahoo = YahooDataSource()

start = datetime(1900, 1, 4)
end = datetime.today()

original = yahoo.get_ticker('IBM', start, end)['Adj Close']
predictions7 = predictARIMA(original.ix[0:-7], 7)
predictions14 = predictARIMA(original.ix[0:-14], 14)
predictions21 = predictARIMA(original.ix[0:-21], 21)
plt.plot(original.ix[-2*50:])
plt.plot(predictions7, color='r', alpha=0.9)
plt.plot(predictions14, color='g', alpha=0.2)
plt.plot(predictions21, color='b', alpha=0.2)
```
